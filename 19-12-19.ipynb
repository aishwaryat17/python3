```python
# operations
#include <stdio.h>

int main()
{
    int a,b,op;
    scanf("%d%d",&a,&b);
    scanf("%d",&op);
    switch(op)
    {
        case 1:printf("%d",a+b);break;
        case 2:printf("\n%d",a-b);break;
        case 3:printf("\n%d",a*b);break;
        case 4:printf("\n%d",a/b);break;
        case 5:printf("\n%d",a%b);break;
        default:printf("\ninvalid input");break;
    }
}

#include <stdio.h>

int main()
{
    int a,b,op;
    scanf("%d%d",&a,&b);
    scanf("%d",&op);
    switch(op)
    {
        case 1:printf("%d",a+b);break;
        case 2:printf("\n%d",a-b);break;
        case 3:printf("\n%d",a*b);break;
        case 4:printf("\n%d",a/b);break;
        case 5:printf("\n%d",a%b);break;
        default:printf("\ninvalid input");break;
    }
}
```

```python
# palindrome
#include <stdio.h>

int main ()
{
  int t,n,r,s=0;
  scanf("%d",&n);
  t=n;
  while(n!=0)
  {
  r=n%10;
  s=s*10+r;
  n=n/10;
  }
  if(t==s)
  printf("palindrome");
  else
  printf("not");
}


# amstrong

#include<stdio.h>
#include<math.h>
void main()
{
 int t,n,r,c=0,s=0;
 scanf("%d",&n);
 t=n;
 while(n!=0)
 {
     c++;
     n=n/10;
 }t=n;
 while(n!=0)
 {
 r=n%10;
 s=s+pow(r,c);
 n=n/10;
}
if (s==t)
 printf("armstrong number");
 else
 printf("not ");
}
```

```python
# amstrong range
#include <stdio.h>
#include<math.h>
int main ()
{
  int t,n,r,s,c,range;
  scanf("%d",&range);
  for(n=1;n<=range;n++)
  {
  t=n;
  s=0;
  c=0;
  while(t!=0)
  {
  c++;
  t=t/10;
  }
  t=n;
  while(t!=0)
  {
  r=t%10;
  s=s+ceil(pow(r,c));
  t=t/10;
  }
  if(n==s)
  printf(" %d",n);
}
}

```
